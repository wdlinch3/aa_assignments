## The Cosmic Distance Ladder

For objects within about 1 AU, distances were accurately determined using radio detection and ranging (RaDAR): We bounce radio-wavelength light off of Venus and wait to hear (see?) the echo to find that it is roughly 0.3 AU at closest approach. 

Stellar parallax is useful for measuring the distance to the nearest stars, but it is useless for anything further out, precisely because we are measuring the parallax relative to the stuff further out. To find the distance to the “fixed stars” we need to find the next rung in this **ladder of cosmic distance scales**. We will do this using the spectroscopic properties of stars, but to talk about those, we first need to discuss the brightness of stars and how it relates to distance.

### Luminosity and Radiant Flux

In Advanced Astronomy: Solar System (AA:SS), we verified the **luminosity** of the Sun to be $L_\odot=3.83\times 10^{26}$ `W` (1 Watt is a Joule per second) and checked that at 1 AU, this gave a solar constant of $F_\odot=1360$ `W/m**2`. The luminosity is a measure of “power” which is the amount of energy produced per unit time (a.k.a. energy production rate). The solar constant is, on the other hand, a measure of power crossing a surface per unit area. In this case, the relevant amount is how much impinges on our upper atmosphere:

> $$\textrm{solar constant} = \frac{\textrm{luminosity}}{\textrm{area}}= \frac{\textrm{energy}}{\textrm{area} \times  \textrm{time}}$$

The luminosity $L$ is an intrinsic property of a star. By contrast, how bright the star appears to us is called the **apparent brightness**. (In the example of the Sun as seen from Earth, this apparent brightness is called the solar constant.) We see here that it is a quantity (energy) *per unit time and per unit area*. In physics, such a quantity is called a "flux". In this case, the apparent brightness is an energy flux called the **radiant flux**. It is usually denoted by the letter $F$.

### The Magnitude Scale

Astronomers use a byzantine measure to describe the apparent brightness of a star. Its origin’s lie in Hipparchus’s classification of stars using a scale from 1 to 6 with 6 being the faintest stars visible with the naked eye (which is all he had). The difference in energy flux between these extremes is a factor of roughly 100, so each step in Hipparchus’s scale corresponds to a decrease in brightness by a factor of $\sqrt[5]{100}\approx 2.5$.

Taking the logarithm of a dimensionful quantity ($d$ has the dimensions of length) does not make sense. We fix this (and the constant $m_0$) by introducing the following normalization: The **absolute magnitude** $M$ is *defined* to be the apparent magnitude an object *would* have *were* you to move it to a distance of 10 `pc`. Putting all this together, we arrive at the relation 
$$
m-M = 5\log \left(\frac d{10 \textrm{pc} } \right)    \tag{1}
$$
between the apparent magnitude $m$, the absolute magnitude $M$, and the distance $d$ measured in parsecs. The quantity $m-M$ appearing on the LHS of equation 1 is a measure of distance called the **distance modulus**.
