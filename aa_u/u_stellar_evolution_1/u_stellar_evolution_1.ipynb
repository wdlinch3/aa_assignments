What we have just shown is that the severely antisocial nature of fermions gives a contribution to the pressure that becomes significant at high densities. In fact, at high enough densities, as with stellar-mass, stage 9 stars, it is this [**electron degeneracy pressure**](https://en.wikipedia.org/wiki/Electron_degeneracy_pressure) that is responsible for stabilizing the core against gravitational collapse. Matter in this state is called [degenerate matter](https://en.wikipedia.org/wiki/Degenerate_matter). It is an inherently quantum-mechanical state of matter, as you can conclude from its dependance on Planck’s constant: If we were to turn off quantum mechanics by taking $h\to 0$ (the so-called semi-classical limit), then the corresponding pressure $P\to 0$.

Notice, however, that contrary to the (relatively insignificant) thermal pressure from problem 4, the electron degeneracy pressure (3) is *independent of the temperature*: An increase in temperature does not give a corresponding extra outward push on the crushing gravitational force of the matter bearing down on the core, so there is no corresponding reaction preventing the gravitational contraction from continuing and correspondingly increasing the core temperature. This is very much *unlike* what we are used to from ideal gasses in which an increase in temperature results in a corresponding increase in pressure according to the ideal gas law (eq. 2). Thus, the temperature eventually rises to the $10^8\ K=$ one-hundred million Kelvin needed to initiate helium fusion at this pressure.

According to table 1, the core temperature of a stellar-mass main sequence star is $T_{core}\sim 1.5\times 10^7\ K$, so the temperature required classically to overcome the Coulomb barrier is too high by a factor of a thousand. Thus, we find that classical physics cannot explain how there can be enough thermal energy in the core of a main sequence star to get protons to fuse into helium: The “Coulomb barrier” is simply too high for protons to be able to get close enough to fuse. 

The resolution of this contradiction is found in the funny business of quantum mechanics: As we have discussed in `u_spectroscopy`, particles are also waves. Waves have an inherent uncertainty in their "position" which allows them to penetrate some barriers that classical particles cannot. Where a classical particle is “at a point”, a quantum particle is within a de Broglie wavelength  of that point. In particular, a quantum proton does not need to be within 1 femtometer of another to fuse, it only needs to be within one de Broglie wavelength of 1 femtometer to have a chance to fuse. This ability to penetrate through classically disallowed barriers is called “[quantum tunneling](https://en.wikipedia.org/wiki/Quantum_tunnelling)”.

When nuclei fuse/fission the total mass before and after the reaction will generally not be equal. This happens because there is “binding energy” required to hold the individual parts together and energy is equivalent to mass by Einstein’s relation $$E=mc^2$$

For example, during the first stage of the hydrogen fusion in the p-p chain, two protons fuse to a deuterium nucleus. A hydrogen ion is a proton $p$ and a deuteron is an isotope of hydrogen with one proton and one neutron. This is possible because one proton $p\to n+e^+ +\nu_e$ is converted to a neutron $n$ by releasing a positron $e^+$ and a neutrino : $$2 {}^1_1H= p+p\to p+n+e^+ +\nu_e\to {}^1_2H+e^+ +\nu_e$$

Such processes are said to proceed by $\beta$-decay. We can calculate the energy gained or lost in this transition by comparing the masses of the particles before and after the reaction. Initially, we have 2 protons each with a mass of $938.272\ MeV$, so a total of $1,876.544\ MeV$. After the reaction, we have a deuteron with mass $1.87561\ GeV$, an electron with mass $.51\ MeV$ and a neutrino with (essentially) no mass. Converting everything to `MeV`, the difference in energy is $$\Delta E=2\times 938.272 -1,875.61 -0.51=0.42\ MeV$$

Let us return to what is happening in the core of the star. As it runs out of hydrogen to fuse in the [**p-p chain**](https://en.wikipedia.org/wiki/Proton–proton_chain), helium accumulates but cannot fuse since the Coulomb barrier is too high (problem 10). The core contracts, converting gravitational potential energy into heat while the hydrogen shell burning speeds up (problem 3.f). The temperature increases, but the core doesn’t respond in its degenerate state (problem 8), so the temperature continues to climb until it reaches the $\sim 10^8$ `K` required for helium fusion (problem 10.f.ii). 

Once helium ignition conditions are reached, helium begins to fuse through the [**triple-$\alpha$ process**](https://en.wikipedia.org/wiki/Triple-alpha_process) (problem 12.c). With the core still in its degenerate state, there is no increase in pressure as the temperature increases even more. (We reiterate that this is very much unlike what we are used to from ideal gasses in which an increase in temperature results in a corresponding increase in pressure according to the ideal gas law (eq. 2). Even when the gas is not ideal, we still have something similar to $P\sim T$, so the behavior of this degenerate matter is very strange.) As the temperature increases, helium burns faster, the temperature goes up, the core does not adjust, so the rate increases again, and so on. This runaway chain reaction–called the [**helium flash**](https://en.wikipedia.org/wiki/Helium_flash)–lasts only a few hours until the temperature has finally started to bring the electrons out of their degenerate state. Once the electrons have been roused from their quantum slumber, the matter returns to its normal state, and the thermal pressure can once again take over, thus bringing the core back into hydrostatic equilibrium. 

## Stage 10: Horizontal Branch

When this happens we enter stage 10. Helium is fusing stably in the triple-α process in the helium core while hydrogen continues to burn in a shell around it. Having blown off up to ~ 20–30% of its mass as a red giant (problem 7), the star lands somewhere in the part of the H-R diagram called the horizontal branch. 

Throughout the journey from main branch to stage 10, the physics is dominated by what is happening in the helium core. This is roughly uniform across stars since they all need to form a helium core and build up to the helium ignition temperature. As long as they are all required to undergo this process, we can be confident that their He III cores are similar. Since the main driver of luminosity is the fusion process in the core, all these stars will end up with similar luminosities, *ergo* the [**horizontal branch**](https://en.wikipedia.org/wiki/Horizontal_branch). Thus, all the stage 10 stars have similar luminosities with the more massive stars on the right (lower surface temperatures) and the less massive stars (with higher surface temperatures) on the left. Altogether, the increased radiation pressure from the interior, the instability of the outer layer, and recombination cause all the material in the envelope to be ejected. Over a span of a few million years all but the core remains, surrounded by a nebula of cool gas (H, He, C, O, Ne) and dust. For historical reasons, the resulting object is called a [planetary nebula](https://en.wikipedia.org/wiki/Planetary_nebula), which is one of the worst misnomers in astronomy. 



The remaining core consists of the left-over carbon and oxygen at a core temperature of $~ 100$ `MK`, a surface temperature of $50,000$ `K` $=50$ `kK`, and a size $\sim R_\odot/100 \sim R_\oplus$ roughly that of the Earth. Due to all of these properties, this object is called a carbon-oxygen [white dwarf](https://en.wikipedia.org/wiki/White_dwarf). All fusion processes have ceased, the gravitational attraction is balanced by electron degeneracy pressure, and–barring additional interactions with other astronomical systems–the white dwarf will proceed to cool by radiation into the near-vacuum of space over billions of years. In theory, the final state should eventually be a [black dwarf](https://en.wikipedia.org/wiki/Black_dwarf). Black dwarfs have not been observed, but this is hardly strange. Firstly, it is entirely possible that the cool-down time turns out to be longer than the age of the Universe, which-as we will learn-is around 14 billion years. (This fact should give you pause: our star will be *on the zero-age main sequence for $10/14\times 100 =71\%$ of this time*! Either our star is in some sense *very* old, or the Universe is in the same sense *very* young!) Secondly, even if black dwarfs did exist, they would be difficult to find because they are so small and so dark. (Recall for comparison that even finding much hotter brown dwarfs and way hotter red dwarfs is a challenge!)
