We call the stuff between the star systems the **interstellar medium (ISM)**. The interstellar medium consists of radiation (a.k.a. light) and interstellar matter. Gravity is an attractive force, so if we start off with a GMC in which all particles are at rest, all the particles will be attracted to their common center of mass and the cloud will contract. If, on the other hand, the particles all have random velocities with large average speeds, the particles can escape each others gravitational attraction and the cloud will disperse.

### The Virial Theorem

In problem 5b of the packet Astrodynamics II, you showed that for the ISS in orbit around the Earth, the (averaged) kinetic energy $K$ and potential energy $U$ satisfied the relation $K=-U/2$.

This is an example of a general rule called the **virial theorem**. This theorem is powerful in the sense that it holds under very general conditions and for very complicated systems. Surprisingly, it is not hard to state andif you know calculussimple to prove. Consider a system consisting of a large number of particles all interacting through a potential $U$. Note that $U$ can be a very complicated function of all the particle positions. Luckily, we will not need to know or manipulate its explicit form. To solve for the dynamics of this $N$-body system means that we would need to find the positions and momenta of each particle at any time. We have discussed how this is already impossible for $N=3$. But suppose we dont want to know exactly where each particle is and what it is doing at any particular time. Suppose instead, we only want to know some gross features of the system averaged over time. 

The ISS example from problem 5b is a simple case of this general situation: Here there is only one particle (N=1, the ISS itself) and we are not asking where it is as a function of time, but only its kinetic energy K. Since the orbit is circular this kinetic energy is constant and so the time average of it is equal to this constant value and we find that the virial theorem holds somewhat trivially. 

For a less trivial example, suppose the orbit of the ISS is on an orbit that is not circular. This means the orbit must be elliptical, and we know from Keplers second law that the speed is not constant. Because of this, the kinetic and potential energies are not constant, so we cannot conclude that K=-U/2.

Now we can see the use of the virial theorem: Although we cannot conclude that $K=-U/2$, the virial theorem says that this relation does hold if you average over the period of an orbit. Let us denote this average by $\ \cdot\ $: The average kinetic energy over a period is $K$ and the average potential energy is $U$. Then the **virial theorem** states that:

> For a *stable* system of any number of particles interacting through a potential $U$ the *time-averaged* kinetic energy and potential energy satisfy
$$
2K+U=0	\tag{1}
$$

### (In)Stability of a Giant Molecular Cloud

Let us apply this to our GMC. In problem 5, you estimated the gravitational potential energy of the cloud, so we conclude that this cloud will be stable if the average kinetic energy of the molecules in the cloud is half of this number. In chemistry, we (will) learn that the average kinetic energy of a (monatomic, ideal) gas is related to the temperature of the gas by the formula 
$$
K=\tfrac32 Nk_B T	\tag{2}
$$
where $k_B$ is called **Boltzmanns constant** and $T$ is the temperature and $N$ is the number of particles. The SI unit of temperature is degrees Kelvin (denoted K) and in SI units, Boltzmanns constant has the value $k_B\approx 1.4\times 10^{-23} J/K$. Boltzmanns constant is a universal constant similarly to Newtons constant $G$ and the speed of light in vacuum $c$.[^2] This constant allows us to convert anything with the dimensions of temperature to something with the dimensions of energy. Since this is always possible, it means that *temperature is just another form of energy*.

9. Use the virial theorem to find the temperature at which your GMC will not disperse and will also be stable against gravitational collapse. Redo your calculations to show that this critical temperature $T_c$ is 
$$
T_c\sim \frac{GM^2}{N_c k_B R}=\frac 1{k_B}\frac{Gm_pM}R	\tag{3}
$$

Let us pause to contemplate this result: If we recombine $k_B T$ we have an energy that is due to temperature; lets call this thermal energy. If the thermal energy is below $k_BT_c=G\frac{m_pM}R$ gravity wins, and the cloud collapses under its own weight. If the thermal energy exceeds this value $G\frac{m_pM}R$, there is not enough gravitational potential energy to keep the cloud from dispersing. Looking closely at this value, we notice that it is the gravitational potential energy between a single proton and the whole rest of the cloud when that proton is separated from the center of mass of the cloud by the radius of the cloud, that is, at the edge of the cloud. In other words, this is the gravitational potential energy you gain when you bring a proton from very far away to the edge of the cloud, that is, when you add another proton to the cloud. If the kinetic energy of this particle is higher than that, it will escape back from the cloud. If not, it will fall into the cloud.

The conclusion from our analysis seems to be that starting from a GMC and waiting a few millennia, we will end up with a collapsed object with a million solar masses. By what we already know of the Schwarzschild radius, this suggests we end up with an enormous black hole. Although many of our calculations were good approximations, what actually happens is somewhat different. To see why, lets take a closer look at our result from problem 10\. 

Going back to your formula for the free-fall time, you should notice that instead of depending on the mass $M$ and radius $R$ of the cloud separately, it only depends on the mass density $\rho \propto \frac M{R^3}$: 
$$
t_{ff}\propto \frac 1{\sqrt{G\rho}}
$$
This implies that as long as the density  is the same throughout the cloud, the result is independent of the size $R$ of the cloud. If the cloud is homogeneous (as we assumed when we took constant density), this suggests that the collapse should be happening evenly everywhere throughout the cloud. But if vastly separated parts of the cloud are all contracting in the same way, that suggests that the cloud will fragment. 

Suppose then that the cloud undergoes this **fragmentation** into smaller clouds that continue to contract. Then instead of an outcome with one super-massive dense object at the center of mass, we might expect a bunch of dense and less massive objects all clustered around a common center of mass. If stars and star systems are to form from this model, the prediction seems to be that these will form in clusters with many star systems coming from one enormous cloud.
