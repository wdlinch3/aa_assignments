We call the stuff between the star systems the **interstellar medium (ISM)**. The interstellar medium consists of radiation (a.k.a. light) and interstellar matter. Interstellar matter, in turn, is mostly gas and dust. The gas can be **ionized**, **atomic**, or **molecular**. ### The Virial Theorem

In problem 5b of the packet `x_astrodynamics_2`, you showed that for the ISS in orbit around the Earth, the kinetic energy $K$ and potential energy $U$ satisfied the relation $K=-U/2$.

This is an example of a general rule called the **virial theorem**. This theorem is powerful in the sense that it holds under very general conditions and for very complicated systems. Surprisingly, it is not hard to state andif you know calculussimple to prove. Consider a system consisting of a large number of particles all interacting through a potential $U$. Note that $U$ can be a very complicated function of all the particle positions. Luckily, we will not need to know or manipulate its explicit form. To solve for the dynamics of this $N$-body system means that we would need to find the positions and momenta of each particle at any time. We have discussed how, in general, this is already impossible for $N=3$. But suppose we dont want to know exactly where each particle is and what it is doing at any particular time. Suppose instead, we only want to know some gross features of the system averaged over time. 

The ISS example from `x_astrodynamics_2` is a simple case of this general situation: Here, there is only one particle ($N=1$, the ISS itself) and we are not asking where it is as a function of time, but only its kinetic energy $K$. Since the orbit is circular this kinetic energy is constant and so the time average of it is equal to this constant value and we find that the virial theorem holds somewhat trivially. 

For a less trivial example, suppose the orbit of the ISS is on an orbit that is not circular. This means the orbit must be elliptical, and we know from Keplers second law that the speed is not constant. Because of this, the kinetic and potential energies are not constant, so we cannot conclude that $K=-U/2$.

#### Jeans instability

Let us pause to contemplate this result: If we recombine $k_BT$ we have an energy that is due to temperature; lets call this thermal energy. If the thermal energy is below $k_BT_c=\frac{Gm_pM}R$ gravity wins, and the cloud collapses under its own weight. If the thermal energy exceeds this value $\frac{Gm_pM}R$, there is not enough gravitational potential energy to keep the cloud from dispersing. Looking closely at this value, we notice that it is the gravitational potential energy between a single proton and the whole rest of the cloud when that proton is separated from the center of mass of the cloud by the radius of the cloud (*i.e.*., at the edge of the cloud). In other words, this is the gravitational potential energy you gain when you bring a proton from very far away to the edge of the cloud, that is, when you add another proton to the cloud. If the kinetic energy of this particle is higher than that, it will escape back to infinity from the surface of the cloud. If not, it will fall into the cloud. 

If this sounds familiar, you might consider rewriting the gravitational energy of the proton in this form: 
$$
\frac{Gm_pM}R=\frac12 m_p \left(\sqrt{\frac {2GM}R}\right)^2
$$

## Cloud fragmentation

However they come about, we have seen that a dense core inside a GMC will typically be unstable to gravitational collapse. We will now focus our attention on the dynamics of this collapse. 

Recall that our cloud starts out enormous with a radius of 80 lightyears. The solar system itself is only 40 AU in radius, and since 1 AU corresponds to 500 light-seconds, the solar system is only 11 light-hours across, or 130,000 times smaller than the GMC. Thus the solar system is basically 0 size compared to the original cloud. If we want to know how long it takes this cloud to collapse, we may therefore say that it is the time it takes to contract from 80 ly to 0 ly. 

Let us also recall that throughout much of this collapse, the cloud has a very low density. Because of this, molecules in the cloud do not collide throughout much of the contraction. Instead, they free-fall in the gravitational field created by all the other molecules.

The conclusion from our analysis seems to be that starting from a GMC and waiting a few million years, we will end up with a collapsed object with a million solar masses. So again, we have pursued solid reasoning and derived a conclusion, and again as always when this happens, it means we have missed some important physics.  To resolve the issue, lets take a closer look at our result from problem 12. 

Going back to your formula for the free-fall time, you should notice that instead of depending on the mass $M$ and radius $R$ of the cloud separately, it only depends on the mass density $\rho\sim \frac M{R^3}$: 
$$
t_{ff}\sim \frac1{\sqrt{G \rho}}
$$ 
This implies that as long as the density  is the same throughout the cloud, the result is independent of the size $R$ of the cloud. If the cloud is homogeneous (as we assumed when we took constant density), this suggests that the collapse should be happening evenly everywhere throughout the cloud. But if vastly separated parts of the cloud are all contracting in the same way, that suggests that the cloud will fragment. 

Suppose then that the cloud undergoes this **fragmentation** into smaller clouds that continue to contract. Then instead of an outcome with one super-massive dense object at the center of mass, we might expect a bunch of dense and less massive objects all clustered around a common center of mass. If stars and star systems are to form from this model, the prediction seems to be that these will form in clusters with many star systems coming from one enormous cloud.
