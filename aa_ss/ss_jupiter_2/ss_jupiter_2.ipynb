The tidal forces we discussed in the previous packet can tear an astronomical body apart if it comes too close to the primary that is causing the tidal forces. Consider again the tidal field you derived there and imagine scaling it up in strength until rocks on the near side start accelerating away from the surface because of the difference in gravitational pull while the rocks on the far side are accelerated away by the centrifugal force. You can work this out precisely, as Roche did in his derivation of the so-called **Roche radius** or **Roche limit**. The formula he found was
<a tag:roche>
<a id="roche">$$r_{Roche}= f_{Roche}\sqrt[3]{\frac{\rho_M}{\rho_m}} R_M  	\tag{1}$$

In this formula, $\rho_M$ is the mass density of the primary, $\rho_m$ is the density of the satellite, and $R_M$ is the radius of the primary. The coefficient $f_{Roche}$ varies depending on how cohesive the satellite is. If it is liquid, it will deform into an oblate spheroid, so the tidal forces on the opposite ends will be larger. Roche calculated that $f_{Roche}\approx 2.44$ for such an object. If the object is rigid, it does not deform and the tidal forces are smaller. Then  $f_{Roche}=\sqrt[3]{3}\approx 1.44$.

In general, there are other forces helping to hold the body together that can change this numerical "Roche coefficient", but for astronomical bodies $f_{Roche}\sim 1$. Note that it is a mistake to think something along the lines of "Jupiter's tidal field is so strong that anything that gets too close will be ripped apart". In fact, the self-gravity of any potential satellite or impactor with a density of $\rho_m > 2\rho_M\approx 2.6$ `g/cm3` would suffice to hold it together, at least until it gets to the "surface" of Jupiter. ### The Trojans

In this section, we will discuss the astrophysics of three bodies interacting through gravity as they co-rotate (*i.e.* all revolve around their common center of mass). Specifically, Jupiter orbits the *much* more massive (by a factor ~1000 found in `ss_jupiter_1`) Sun at a large distance and satellites much less massive than Jupiter orbit it at radii that are *much* less than the distance to the Sun. 

Of course we could study this system in the inertial frame of reference “looking down” on the Sun-Jupiter-satellite system from the “top” with the Sun roughly sitting still and Jupiter orbiting it (while being orbited itself by the satellite). But to understand what things look like from the point of view of someone near Jupiter, we should instead jump to the *non-inertial* frame revolving at Jupiter’s orbital speed. Recall that when we do that, Newton’s second law gets Coriolis and centrifugal corrections as in equation 1 of `ss_jupiter_1`. More interestingly, note that the direction of the new term is away from the central mass. ### Potential energy surface

A useful visual device is called an “potential energy surface”.  In physics we learn that we can associate an energy to the gravitational interaction and write a potential energy function for gravity. When there is more than one gravitating object, you simply add all their potential energy functions together. If you plot the surface, higher points correspond to places of higher gravitational energy and lower points correspond to lower energy. Maxima and minima of the surface correspond to points where the forces are balanced. Because of this, the energy surface gives a visual representation of how the forces act at all points in space. For Jupiter’s orbit of the Sun the effective gravitational potential energy is schematically of the form
$$u = - \frac 1a - \frac mb - r^2 \tag{3}$$

where $r$ is the radial distance from the Sun and $m=\frac{M_J}{M_\odot}$ is the ratio of Jupiter’s mass to the Sun’s. The $a$ and $b$ are functions of the radius and the angle $\theta$ in the plane from the line connecting Jupiter to the Sun. Explicitly, $a= \sqrt{ m^2+r^2 +2rm \cos\theta}$ and $b= \sqrt{ 1+r^2 -2r \cos \theta}$.

### Maxwell speed distribution

The molecules that make up the gas do not all have the average speed you calculated above. The distribution of speeds in an ideal gas is called the **Maxwell velocity distribution**. As you will learn in AMT, it is of the form 
$$
f(v)=a^{\frac 32} v^2 e^{-av^2} \quad \quad \textrm{where}\quad \quad  a\propto\sqrt{\frac\mu{T}}
$$

You can see from the Maxwell distribution that there are molecules in the gas with speeds much higher than the average velocity. In fact, there is always a small but non-zero fraction of the molecules that have speeds higher than the escape speed for that astronomical body. When these molecules escape into space, the continual random collisions of the remaining gas replenishes this high-speed “tail” of the distribution by giving some molecules that had lower speeds higher speed. This process will continue as long as the temperature of the gas is not lowered, so any atmosphere will gradually leak into space. How long it takes for a substantial part of the atmosphere to escape in this way depends only on the ratio between the average molecular speed (so temperature and molar mass) and escape speed (so mass of and radius of astronomical body). For an atmosphere to stick around for billions of years, the rule of thumb is that $v_{ave}$ should be at least a factor of 6 smaller than $v_{esc}$.
