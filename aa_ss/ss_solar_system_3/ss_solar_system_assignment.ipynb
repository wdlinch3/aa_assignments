# Solar System III: Nebular Theory

In this packet, we finish our discussion of the **solar nebular disk model**. Recall that we studied the fate of a giant molecular cloud collapsing and fragmenting under its own gravity: When the mass of the cloud of temperature $T$ and mass density $\rho$ exceeds the Jeans mass 

$$
M_J=\frac 1G\left( \frac{k_BT}{m_p} \right)^{\frac32} \frac1{ \sqrt{G\rho} }   \tag{1}
$$
the temperature of the cloud, and therefore, the average kinetic energy of the molecules is insufficient to prevent the collapse. The collapse depends only on the density of the cloud and the contraction is the same in all parts of the cloud, leading to fragmentation. Each fragment collapses to its own **proto-stellar nebula** and one of them becomes our **proto-solar nebula**. We note that this picture suggests stars are born in groups, and this is indeed true. Because of this, the nebulae in which star birth is happening are often called **stellar nurseries**. 

We now focus on our proto-solar nebula. For simplicity, let us pretend that our nebula is roughly spherical. All the particles making up this nebula have random velocities of various speeds and in various directions. In a `x_astrodynamics_2`, we learned that the average of these gives the velocity of the center of mass of the nebula. For simplicity, again, we will go to an inertial frame of reference in which we are traveling along with the cloud. When we do this, the center of mass is at rest from our point of view. Now, all the particles will be orbiting this center of mass on randomly-oriented orbits, and we can average over orbits. Although the orbits are randomly distributed, statistically, we will always have slightly more orbits in one plane than another, so when we average we will find that there is a plane through the center of mass that has more orbits roughly parallel with it than any other plane. Furthermore, with respect to this plane, slightly more or fewer orbits will be directed clockwise *versus* counter-clockwise. The result is that even though individual particle motions look random, overall the nebula has an average direction of rotation. 

Just as it is difficult to change the “quantity of motion” (linear momentum) of an object with a large “quantity of matter” (mass), so too is it difficult to change its “quantity of *rotational* motion”. I have no idea whether Newton called this property “quantity of rotational motion”, but the modern term is **angular momentum**. Angular momentum is most often denoted by the letter $L$. Just as you need a force to change a linear momentum (Newton II), so too a “twisting force” is needed to change angular momentum. The “twisting force” is called **torque** denoted by the Greek letter $\tau$ (tau). We will not need to know what torques are exactly, but we can reason by analogy. Just as forces change linear momenta, so torques change angular momenta:
$$
F_{net}=\frac {dp}{dt}\quad (\textrm{linear})\quad versus\quad  \tau=\frac {dL}{dt}\quad (\textrm{angular})
$$ 
Returning now to our slightly-rotating cloud of molecules, we are saying that this large cloud has some angular momentum $L_i$ where the $i$ stands for “initial”. When there are no external forces acting on a system, then the linear momentum of the system is a constant in time. Similarly, since there are no external twisting forces acting on our cloud, the angular momentum of the cloud is also constant in time. Remember, though, that the cloud is contracting. If we wait until the cloud is smaller, it will have an angular momentum $L_f$ where $f$ stands for “final”. Constant angular momentum means that the big cloud’s angular momentum is the same as that of the small cloud. In symbols: 
$$
L_f=L_i  
\hspace{20mm}
\textrm{(angular momentum conservation)}
$$
Actually, we already learned this when we stood on a spinning iron table and pulled our arms in. (We also did something similar with a bowling ball.) The angular momentum with your arms out is $L_i$ whereas the angular momentum with your arms in is $L_f$. When you had your arms out, it was more difficult to spin you up than when you had your arms down: Your **rotational inertia** went from high to low. This concept or rotational inertial is the rotational analog of mass: It's difficult to change the velocity of a large mass. Similarly, the rotational analog of velocity is called **angular velocity**: Just as the velocity is the rate of change of position, so the angular velocity $\omega$ (omega) is the rate of change of angle $\theta$ (theta): 
$$
v= \frac{dx}{dt} \quad (\textrm{linear})\quad versus\quad \omega=\frac{d\theta}{dt}\quad (\textrm{angular})
$$

Coming back to our collapsing cloud, the large cloud has some initial angular momentum $L_i$ but since the cloud is so large, it is barely noticeable that it has any rotation. Nevertheless it *is* rotating in a plane through the center of mass defined by this angular momentum as described in reference to the average particle orbits above. The particles in the cloud are attracted to the center of mass, so they fall toward it, but in doing so, they must maintain the overall angular momentum (since it is constant). The only way this can happen is if they start to orbit faster. Because they are spinning, they feel like they are being flung outward the same way a salad spinner removes water from lettuce, and it is difficult for them to reach the center. Because of this, the giant cloud deforms into a disk-like distribution of particles that is spinning at a noticeable rate. As the cloud continues to contract, it continues to **spin up** and flatten out, so that in the final stages of the collapse we are left with a giant spinning pancake of gas and dust. This is the **proto-solar disk**.

## Planetary Formation: Condensation, Accretion, and Coalescence 

Instead, let us turn our attention to the rest of the fledgling solar system. Recall that our giant molecular cloud had trace amounts of dust strewn throughout a sea of hydrogen and helium ions and molecules. These grains of dust give rise to two important effects: 
- Firstly, they absorb and reradiate light, thereby thermally cooling the cloud, lowering its temperature to below the critical temperature for gravitational collapse.
- Secondly, they act as seeds for **condensation**, in a process very similar to how raindrops form.[^1] When the temperature of a cloud drops, the average kinetic energy of atoms and molecules drops. When their speeds are low enough, an encounter with a dust molecule may cause them to bind with it by the van der Waals (electrostatic) force. Thus, dust grains are **condensation nuclei**. Once such clumps form, they encounter more material randomly and bind to it electrostatically making them bigger. Before eventually, their mass causes a slight gravitational attraction leading to more than just random encounters, and they grow larger. More mass means more attraction, so more mass, *et cetera*. This is the process of **accretion**. 

<div style = "text-align: center"><img src="./Images/dust.jpg" width="500"></div>

To recapitulate: We started with a **giant molecular cloud** (**GMC**) that collapses under its own weight because **interstellar dust** effectively radiates thermal energy from the cloud and lowers its temperature to a point that the gravitational potential destabilizes it. As the cloud begins to collapse, it does so by **spinning up** to preserve **angular momentum**. The outer parts of the cloud collapse to a disk and suffer only a modest amount of heating because they are diffuse. Meanwhile the core of the cloud becomes very dense, trapping heat and increasing in temperature by hundreds of thousands of degrees as it becomes a **protostar**. 

While this is happening, the rest of the cloud is undergoing the processes of **condensation** and **accretion** to form **planetesimals** from the ISM. These are all traveling in similar orbits allowing them to **coalesce** further into **protoplanets**. 

## Planetary Differentiation

Recall that in order for a molecular cloud to be able to contract, it had to be able to radiate away heat to lower the temperature so that the kinetic energy falls below the gravitational potential energy at the onset of the Jeans instability. This is possible because the density of the cloud is low, so the light carrying away the heat can escape. 

As the cloud contracts, the gravitational potential energy is transformed into kinetic energy of the particles in the cloud which is radiated away by the escaping light. The result is that as the cloud contracts, the temperature of the cloud stays the same. This is called **isothermal contraction**. 

Eventually, however, the density in the core of the cloud grows to a point that that light cannot easily escape. Now the gravitational energy is converted into kinetic energy that cannot be radiated away, but instead, heats the particles in the cloud. The molecular cloud retains this increased kinetic energy that is being converted from the gravitational energy of the contraction, so the temperature of this part of the cloud increases. At this stage the heat in this part of the cloud is nearly constant, and the name for this is **adiabatic contraction**. 

Overall then, the temperature at the core of the **protoplanetary disk** grew by many orders of magnitude, but this was not true for regions further out. Detailed simulations indicate that the temperature profile is well-fit by the power law
$$
T=T_0 \left( \frac{r_0}r \right)^\frac34 	\tag{2}
$$
where $T_0$ and $r_0$ are some positive constants.

As you just noticed, the ability of materials to form depended on their distance from the proto-sun. So called [refractory materials](https://en.wikipedia.org/wiki/Refractory) are resistant to decomposition by heat and can form much closer to the center of the proto-planetary disk than more volatile materials. The [volatiles](https://en.wikipedia.org/wiki/Volatile_\(astrogeology\)) can only form much farther out. This distribution of condensing materials gives rise to **planetary differentiation**: the phenomenon that the fraction of refractory materials in the plantes decreases as we move farther out from the Sun, and that–conversely–the fraction of volatiles increases. It explains why the terrestrial planets are denser than the jovians, why Mercury has more heavy elements than Mars, why the  Jovians are almost completely made up of volatiles, *et cetera*. 

### Gravitational Assists

One way to get around in space is to use rockets, but fuel is finite, so we try to minimize the amount burned to change momentum. An efficient way to gain momentum without using fuel is to slingshot around a large body that already has a lot of momentum. This maneuver is called a **gravitational slingshot** or a **gravitational assist** (see Discovery 6-1 on p 142). For example, the [Europa Clipper](https://en.wikipedia.org/wiki/Europa_Clipper), a mission to Jupiter’s fourth-largest moon, will spend the two years flying to Mars, slingshotting around, catching back up to Earth and slingshotting around Earth to finally be on its way to Jupiter (which it will reach a little over three years later).

Let us see how gravitational assists work in a simplified setting pictured below. 

<div style = "text-align: center"><img src="./Images/gravity_assist.png"></div>

Here we have a large body with momentum $P$ and a small one with momentum $p$. The small one will slingshot around the large one and come back the other way with momentum $p'$. The large body will also have a slightly different momentum $P'$ because of Newton’s Third Law, but if it is much more massive than the smaller object the new momentum will be very nearly the same as the old one $P'\approx P$. 

As you will learn in Physics, this type of process is an example of a phenomenon called a “collision” (even though the two bodies never touch). You will also learn that momentum is always conserved in a collision, but *kinetic energy* may or may not be. (When kinetic energy is conserved–as it is here–the collision is called elastic.) Conservation of momentum means that 
$$
P-p=P'+p'	\tag{1}
$$
The minus sign is there because momentum is a vector and I’m picking the direction to the left to be positive. Conservation of *kinetic* energy means that 
$$
\frac{P^2}{2M}+\frac{p^2}{2m}=\frac{P^{\prime 2}}{2M}+\frac{p^{\prime 2}}{2m}	\tag{2}
$$
(Now there is no minus sign because energy is a scalar not a vector.) Here $M$ and $m$ denote the large and small masses, respectively. Hopefully, you will learn in your Physics class, how to solve equations (1) and (2) for the speeds $v'$ and $V'$ after the collision in terms of the speeds $v$ and $V$ before the collision. The solution is 
\begin{align}
v'&=\tfrac{M-m}{m+M}v+\tfrac{2M}{m+M}V \cr
V'&=\tfrac{M-m}{m+M}V-\tfrac{2m}{m+M} v	\tag{3}
\end{align}

Although we have only discussed the one-dimensional case, the principle of large bodies imparting significant momentum to smaller ones that venture too close is a general one. Because the larger body is so massive, even a miniscule change in its velocity translates to an enormous change in its momentum. Since momentum is conserved in the interaction, this large momentum must be transferred to the smaller body. Since the smaller body has such a small mass, this large momentum transfer results in a large change in velocity.

Besides using this as a spacefaring trick, instances of this gravitational slingshot are important to the structure of the solar system. Planetesimals that happened to pass too close to a gas giant were imparted with various amounts of linear and angular momentum. Many were flung completely clear of the solar system and now make up the Oort cloud of icy comets on orbits with a large range of inclinations. Others were flung inward toward the terrestrial planets giving rise to a mechanism by which vast amounts of water may have arrived on these inner planets. (Recall that water could not condense in the inner solar system because the snow line was out around 5 AU.)
