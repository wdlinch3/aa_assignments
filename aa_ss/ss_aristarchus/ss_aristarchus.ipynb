Aristarchus of Samos (310-230 BCE) was probably the greatest astronomer of all time. Unfortunately, he was also ahead of his time—way ahead of his time. Our ability to measure the sizes of things and the state of mathematics at the time were insufficiently developed for the prescient natural philosophy of Aristarchus to have shaped history. Instead, we would spend the next 2000 or so years rediscovering what he was saying all along.

<div style = "text-align: center; font-style: italic; font-size: 12px">Fig1: A diagram taken from Aristarchus’s <a href = https://en.wikipedia.org/wiki/On_the_Sizes_and_Distances_(Aristarchus)> On the Sizes and Distances (of the Sun and the Moon)</a></div>

Specifically, Aristarchus:
* put the Sun at the center of the known universe,
* arranged the 5 known planets in the correct order,
* stated that the Earth **revolves** around the Sun in one year and **rotates** on its axis in one day,
* believed that the stars were like the Sun, only much farther away (and that that is why they look like they are fixed in the sky *i.e.* not displaying noticeable parallax),
* pointed out that the Sun and the Moon have the same angular size,
* calculated the sizes of the Sun and the Moon relative to that of the Earth,
* calculated the distances to the Sun and the Moon in terms of the size of the Earth.

Let us emphasize that this was almost <ins>*two millennia*</ins> before the **[Copernican Revolution](https://en.wikipedia.org/wiki/Copernican_Revolution)**. Some of what is on this list is nothing more than (correct) hypothesis, but some of it is hard science and math. Let’s see how he determined some of these important facts.

> ### Uncertainty *versus* error
> This is a good time to pause to contemplate error and **uncertainty** in astrophysics. As we saw in the problem above, even small uncertainties in certain measurements can give large uncertainties in the values of quantities derived from them. It is a popular convention to shorten “un-cer-tain-ty” to “err-or”, but *no mistake* is being made here. For any particular measurement device, it is simply not possible to determine the measured value to a precision higher than that allowed by the limits of the device itself. When describing measurements, we will take **error** to be synonymous with measurement uncertainty.

Now, if only we could figure out the size of the Earth and its distance to the Sun, we’d be all set. More than half a century after Aristarchus’ death, Eratosthenese figured out a method to determine the size of the Earth. A diagram representing his logic is shown below. 

Eratosthenese noticed that on a certain day at high noon, the light from the Sun illuminated the bottom of a deep, narrow, vertical well in Seyene. This meant that the light was coming in perpendicular to the surface of the Earth at that specific date, time, and location. Similarly, in his travels, he had noticed that on the same day of the year (but different year) at the same time of day, a tall vertical obilisk in Alexandra cast a shadow. Since the obelisk is vertical, this implies the light coming from the Sun is not perpendicular to the surface of the Earth in Alexandria. But this only makes sense if the Earth is not flat. He further realized that if the Earth was a perfectly spherical ball, he could figure out its radius if he measured the angle of the shadow and the distance between Seyene and Alexandria.

## Summary

We learned that from elementary geometry, we can determine 
* the *ratio* of the Earth-Sun and Earth-Moon distances (using the angle between the Sun and the Moon in the quarter-moon phase),
* the *ratio* of the diameter of the Sun to that of the Moon (using the coincidence that they are the same angular size as evidenced by total eclipses),
* the *ratio* of the diameter of the Sun to that of the Earth (using the radius of curvature of the shadow on the Earth on the Moon during a partial eclipse), and
* the size of the Earth (using Eratosthenese).
With this information, we can further determine absolute sizes of the three bodies, but we cannot determine their absolute distances. (This will have to wait until Kepler, which I remind you is almost two millennia after Eratosthenese.)

We also learned that this process is *very sensitive* to the uncertainties in the measurements of certain angles. This means that if angles can only be measured to an uncertainty of an arcminute or so, this will generate uncertainties in our distance ratios of ~15%. These, in turn, give similar uncertainties in the values of the size ratios, *et cetera*. 

Because of the technological limitations on our measurement devices, this "uncertainty" is inherently unavoidable in any quantitative science. For example, according to this [website](https://ssd.jpl.nasa.gov) of Caltech's Jet Propulsion Laboritory (JPL), the mean radius of the Earth is $R_\oplus=6371008.4 \pm 0.1\ m$. This is a spectacularly specific value given that the uncertainty is $10\ cm$.
