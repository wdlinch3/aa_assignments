## Waves

As astronomers, the information we get from the Universe comes to us almost exclusively in the form of light. Since almost everything we know about the cosmos is gleaned from the light we receive, we need to understand light and the way it interacts with matter in the most profound detail we can muster. 

Light is a wave disturbance of the **electromagnetic field**. Freely-propagating electromagnetic waves are also known as **radiation**. To understand what these abstract concepts mean, we will first look at more tangible and familiar examples of waves. Some of the most familiar waves are surface waves of water. Later in life, we may notice waves on strings and we learn that sound is a pressure wave. All the simpler waves have something that they disturb as they **propagate**. This material is called the **medium**. Water waves disturb what would otherwise be a flat and horizontal surface between air and water. Waves on a stretched string disturb the straight resting position of the string. In matter, atoms and molecules are separated from one another by distances that are very close to their average separation. Sound waves are disturbances of these atoms and molecules away from these average separations. 

Most physical waves are very complicated. For example, water waves near a beach change speed and shape as they move and often crest and crash into surf as they approach the shore. However, there are also very simple waves that are quite regular. The simplest waves undulate (crest and trough) in regular intervals in time called the **period** $P$ of the wave. You can make such a wave in a bathtub by taking a ruler and bobbing it up and down regularly at the surface of the water. Straight wave fronts parallel to the ruler will move away from the ruler at a constant speed, and–provided you are bobbing at regular intervals–the distances between successive peaks will be the same for all successive peaks. This distance is called the **wavelength** $\lambda$ (lambda) of the wave.

The reciprocal of the period $f=1/P$ is called the **frequency** of the wave, but it is often more convenient to define the **angular frequency** $\omega=2\pi f=\frac{2\pi}P$. The regular bobbing that you are doing with the ruler is called **simple harmonic motion**. An object undergoing simple harmonic motion is also said to be undergoing regular **oscillation**. If there is a rubber duckie floating on the water then, when the wave reaches it, it will start to oscillate regularly. How much it bobs up and down is called the **amplitude** $A$ of the wave. Note that the *amplitude is independent of the wavelength*: the wavelength is the distance between successive crests and the amplitude is the difference between the height of a crest and a trough.

In problem 1e, you determined that the speed of the wave is related to the period and wavelength by $v=\lambda f$ where $f=1/P$ is the frequency. Although there is nothing incorrect about your result, I find it misleading to leave it as written. As we have discussed, the speed $v$ of the wave is *set by the medium*, not by the relation between wavelength and period. Instead, I prefer to write the formula in this way:
<a tag="dispersion">
$$
\lambda =v \frac 1f \quad\quad\textrm{or} \quad\quad f=v \frac1\lambda
\tag{1}
$$
This way of writing the formula emphasizes that the wavelength of a wave is determined by its frequency (or frequency in terms of wavelength). Such a formula is called a **dispersion relation**.   

To summarize, a simple wave $\Psi(x,t)$ in one space dimension $x$ and one time $t$ can be put in the form   
<a id="wave">
$$
\Psi(x,t) =A\cos (kx - \omega t) \tag{2}
$$
This is a slight rewriting of your result in problem 1. In this expression:
* $A$ is the amplitude.  
* $\omega=2\pi f=2\pi /P$ is the aforementioned angular frequency. The 2 is there because the periodicity of the cosine is $2\pi$.  
* Similarly, $k=2\pi/\lambda$ is called the **wave number**. Up to a factor of $2\pi$, it is the reciprocal of the wavelength.
* The function $\Psi(x,t)$ is called the **wave function**.

## Intensity *vs.* pitch: amplitude *vs.* frequency

An effect that is often mistaken for the Doppler effect in the model above is the *intensity* (loudness) of the sound rather than the *pitch* (frequency): When an object that is making a sound is far away from us, the sound has a lower intensity than when it is closer to us. Suppose the noisy object is emitting sound in all directions evenly. This means that as the sound propagates outward, it is being spread evenly over a larger surface area. If the intensity of the emitter is constant at the source, that means that the intensity perceived as we move away from the source should fall off as $1/\textrm{distance}^2$. We can refine our model so that it takes this new effect into account as well.

In our derivation of the Doppler formula, we used sound waves as an example because they are more familiar to us, but our derivation was independent of the type of wave. Since light is a wave as well, we might ask whether it too suffers a Doppler shift if there is relative motion between the observer and the source. This is one of the most important ideas in observational astronomy: if we know what type of light something is emitting, then comparing the wavelengths that we *observe* to what they *should* be, we can determine the component of the velocity of the source directed toward/away from us.

> b. When we observe this object, we find instead that $\lambda'=460$ `nm`.

>    - iii. What is the speed of the source? What fraction of the speed of light (in vacuum) is this?

<img src="./images/spectrum.png" alt="spectrum" width="150">


YOUR ANSWER HERE

```python
from aa_tools.astroconst import spectrum_wavelength
from astropy.units import nm, m, km, s
from astropy.constants import c

_lamda=560*nm
_lamda_prime=460*nm

print(f"\na.     From the figure, a wavelength of {_lamda} is {spectrum_wavelength(_lamda)}.")
print(f"\nb.     From the figure, a wavelength of {_lamda_prime} is {spectrum_wavelength(_lamda_prime)}.")

redshift = lambda x, x_prime: (x_prime/x-1)*c

_blueshift=redshift(_lamda.to(m), _lamda_prime.to(m))

print(f"\nb iii. An object emitting at wavelength {_lamda} but apparent wavelength {_lamda_prime} has a\n       radial velocity of { (_blueshift.to(km/s)).value:.1e} {(_blueshift.to(km/s)).unit}.")
print(f"       This is {round(_blueshift/c, 2)}c.\n")
```

In the problem above, the light’s wavelength is shifted to the shorter side. In the visible part of the spectrum, blue light is on the short side (450 – 485 `nm`) whereas red light is on the longer side (625 – 750 `nm`). Because of this, we refer to light that is Doppler shifted to shorter apparent wavelengths as **blue-shifted** and light that is Doppler shifted to longer apparent wavelengths as **red-shifted**. We use this terminology not just for light, but for other waves as well. 

Note that the Doppler shift we have been discussing can only tell us about the part of the relative motion that is directly toward/away from us. This type of motion is called **radial motion**. (The transversal component of the motion is called **tangential motion** (because this component is tangent to the celestial sphere), but the Doppler shift as we have discussed it, does not tell us anything about that part.)
