## The Rutherford model

In AA:SS, we learn about how planets orbit the Sun: Kepler's First Law states that around a large central mass $M$, a planet of mass $m\ll M$ orbits in an ellipse with the large mass at one of the two foci of the ellipse. Newton derived this law from his laws of dynamics (Newton I-III) and the Law of Universal Gravitation:
$$
F=G\frac{mM}{r^2}
$$

Most orbits "circularize" so that it is usuall a good approximation to take the orbit to be a circle. In this approximation, we show in `x_astrodynamics` how to derive Kepler's laws from the fact that an object in circular orbit has a constant, non-zero centripetal (radially inward) acceleration with magnitude $a=v^2/r$. 

In chemistry and physics, we learn about Coulomb's law for electric charge. If we have two object carrying charges $q_1$ an $q_2$, then there is a force between them with magnitude 
$$
F=k\frac{q_1 q_2}{r^2}
$$
where $k$ is Coulomb's constant. To talk about charge, we need to have a way of talking about "how much charge". In honor of Coulomb, the unit of charge was named the **Coulomb** and abbreviated as `C`. This is still the SI unit of charge, and in these units, the Coulomb constant is approximately $k\approx 9.0\times 10^9\ \frac {kg \cdot m^3}{s^2 \cdot C^2}$. Note the massive disparity in the order of magnitude of the constant relative to Newton's constant in SI units. Although we should not compare apples and oranges to closely, this is an early indication that the electric force is in some sense enormous compared to the gravitational force.

There seems to be quite a strong analogy between Coulomb's law and Newton's law of universal gravitation, but there are some very big differences. For one thing, charges can be positive, negative, or zero, whereas masses are always positive. Like charges (both positive or both negative) repel. Opposite charges (one positive and one negative) repel. By contrast, Newton's gravitational force is always attractive. (If that bothers you, come talk to me.) Another difference is that mass plays two roles: it is the resistance to a change in motion, but it is also the source of a gravitational field. For charge, on the other hand, there is no analogous inertia. You can have as much charge as you want and you won't be any harder to push around. 

The symbol for the *magnitude* of the electron charge is $e$, so the actual electron charge is $-e$ and the charge of the proton is just $e$. Why don't we call it $p$ then? I think we do it this way, because it is a pain for fundamental constants to be negative and we don't usually call it the proton charge, because the electron is in a certain sense more "fundamental". The whole negative/positive mess actually came from the fact that early electrophysicists assumed the charge carriers in metals were positively charged. (This is also why the conventional current direction is opposite to the flow of electrons.)

Since all the ordinary matter in everyday life is made of proton, neutrons, and electrons, any charged object must have a charge that is an integer multiple of the proton charge. We call it the "fundamental" charge for this reason. (Of course for macroscopic objects you cannot tell that charge is not continuous.) For any object of charge $q$ there is a (usually enormous) integer $n$ for which $q=ne$ holds. This empirical fact is called **charge quantization**. 

We want to use Coulomb's law to build a model for how an atom works. In the simplest case, this would consist of a "bound state" of an electron and a proton. The proton mass $m_p\gg m_e$ is much larger than the electron mass, so we expect it to be a good approximation to think of the proton as fixed while the electron can move. This is very similar to the Kepler problem, so perhaps the electron orbits the proton in an elliptical motion or circular motion. This model is called the Rutherford model of the atom. 

The Rutherford model of the atom is wildly incorrect. For one thing, an electron in orbit around a proton has a non-zero acceleration at all times. But we learned in `x_waves_and_light` that accelerating charges radiate light. Thus, an electron in orbit radiates light and that light carries away energy, so the electron spirals into the proton. Since the electric force is so large and the mass of the electron is small, the electron loses energy at a ridiculous rate and the Rutherford model is not even the tiniest bit stable. As with the ultraviolet catastrohe, classical reasoning has failed spectacularly to explain what is going on with atoms.

## Quantum Mechanics

It's actually surprising that all of this pseudo-quantum semi-classical reasoning works at all. In our modern understanding of Quantum Mechanics, the electron is described by a complex-valued function $\Psi(t,x,y,z)$ of space and time called the *wave function*. Since it is complex, it cannot describe a physical property of the electron directly. We might think of its modulus as describing where the electron is, but an electron is particle so half cannot be in LA while the other half is in NY. The prescription that works is that the squared modulus $|\Psi|^2$ can be interpreted as a *probability density* of finding the particle. More precisely, the probability $dP(t,{\bf x})$ between time $t$ and a nearby time $t+dt$ in a box at $\bf x$ with volume $dxdydz$ is 
$$
dP(t,{\bf x})=|\Psi(t,{\bf x})|^2\ dxdydzdt 
$$
If $|\Psi|^2$ is a sharply-peaked Gaussian, then the position of the particle is pretty well-defined, but otherwise the uncertainty in where you should look to find it is large. How the wave function evolves over time is determined by the Schrödinger equation. I'll write it here just for fun
$$
i\hbar \frac{\partial \Psi}{\partial t} = -\frac {\hbar^2}{2m}\nabla^2 \Psi +U({\bf x})\Psi
$$
where $U({\bf x})$ is the classical potential energy. 

If you wanted to derive the results we did by making plausible conjectures properly, you would take the following steps:
- Plug in the Coulomb potential from problem 3 in for $U$.
- Use the separation of variables trick to simplify $\Psi(t,{\bf x})=e^{-\frac{iEt}\hbar}\psi({\bf x})$. Then the Schrödinger equation reduces the simpler time-independent Schrödinger equation (TISE) which looks like the original but with the RHS replaced with $E\Psi$. $E$ is a constant with the interpretation of the total mechanical energy (a concept from physics). In fact, the TISE is just the statement $E=K+U$ realized on $\Psi$. 
- Switch to polar coordinates and try separating variables again to get a radial equation, and two angular equations from the TISE
- Take a few classes to learn how to solve these equations in terms of special functions called Hermite polynomials and spherical Bessel functions.

### Absorption

The process just described is called **spontaneous emission**. In quantum mechanics, all processes that preserve the norm of the wave function are called *unitary*. (Unitarity is needed as a condition because we want to interpret $|\Psi|^2$ as a probability density, and we cannot do that if the normalization is changing.) A unitary process can be undone, and this turns out to mean that if a process is allowed in quantum mechanics, then the reverse process is also allowed and will have the same probability: Quantum mechanics is time-reversal invariant. 

The time-reversal of an emission event would be the absorption of a photon to make an electron transition from the low-energy state to the high-energy state. Therefore, if we pass white light though a gas, we expect the spectral lines to be *subtracted*. This is called an **absorption** spectrum. Hot gasses emit in a characteristic spectrum and cool gasses absorb in exactly the same spectrum. These two rules had been established empirically by Kirchoff before the development of quantum mechanics, so they are called Kirchoff's laws by astronomers. (For a physicist, Kirchoff's laws usually refer to a completely unrelated set of rules for circuits.)

### Line broadening

The analysis we have just completed suggests that the energy of an emitted photon is a very specific number $E$ and therefore a very definite wavelength $\lambda$. In practice, however, this process is subject to various effects that broaden the line. At the most fundamental level, quantum mechanics itself implies a limiting minimum spread $\Delta \lambda$ in the wavelength. Consider, for example, an unstable excited state that transitions to a lower-energy state. This state has a "lifetime" $\Delta t$ which is the timescale an excited electron will stay in that state before it transitions.  Then quantum mechanics dictates that there is an associate uncertainty in the energy of the transition that is bounded below by $\Delta E\lesssim \frac \hbar\Delta t$ where $\Delta t$. Since $E=hc/\lambda$, this implies that there is an inherent uncertainty $\Delta \lambda$ in the wavelength. 
$$
\Delta \lambda \simeq \frac{\lambda ^2}{2\pi c}\frac 1 {\Delta t} \tag{$*$}
$$


More importantly, there are other effects either coming from the motion of the molecules or interactions disturbing the process and changing the state of the electon while it is in the process of emitting a photon. 

For the former, consider a hot dilute gas of particles at some fixed distance from you. If the distance is fixed, the gas has no velocity relative to you. Since the particle motion in the gas is random, half the particles have a velocity component toward you, and half have a component directed radially away from you. In other words, half the particles are blueshifted and the other half are redshifted ever so slightly around the true transition frequency, causing a spread in the line. This phenomenon is called **Doppler broadening** of the line.

Finally, we have the **Zeeman effect**. Elementary particles like electrons and protons have an intrinsic property that is deeply connected to how they behave under rotations. For example, a "vector" particle behaves the way an arrow does when you rotate it. For example, taking an upward-pointing vector and rotating it through an angle of $\pi$ the right way will make it point in the downward. There are generalizations of vectors that rotate in ways that are different from this. How a particle behaves under rotations is an *elementary property intrinsic* to each elementary particle called the **spin** of the particle and denoted by $s$. A spin-$s$ particle has an immutable angular momentum $L=s\hbar$ as one of its defining properties.

> Congratulations! You have just learned the last of the fundamental properties of particles! Any fundamental particle in nature is uniquely identified by its mass $m$ it's charge (in units of fundemential charge $e$) and its angular momentum in units of the reduced Planck constant $\hbar =\frac h{2\pi}$. Not that two of these quantities are discrete. It is currenly unclear whether the mass is also discrete, but there are good reasons to believe that they are multiples of the Planck mass $m_{Pl}=\sqrt{\frac{G\hbar}{c^5}}$. (If you already took AA:SS, you should instinctively be checking units at this point.)

Just as moving charges give rise to magnetic field, so too do spinning charges. An electron has carge $-e$ and the smallest non-zero amount of spin $s$. For example, the spin of a photon is $s=1$ (it is a vector particle as we learn in a class on electromangetism), and electrons, protons, and neutrons all have spin $s=\frac12$. For deep mathematical reasons, spin is "quantized" like charge, but unlike charge, it can carry half-integral values as well as integral values. 

It will therefore have a small immutable magnetic field called its **magnetic dipole momenent**, often denoted by $\mu$. Classically, the magnetic moment of a spinning ball of mass $m$, charge $q$, and spin $s$ is given by $\mu = \frac {g q\hbar}{2m} s$ with the "$g$ factor" $g=1$. When relativistic quantum mechanics is taken into accout, we find instead that for fundamental particles like electrons and muons, $g=2$. The magnetic moment of an electron in quantum mechanics is, therefore 
$$
\mu_B= \frac em \frac \hbar2
$$
which is a quantity called the **Bohr magneton**.

The reason this is important for us is that stars often have a sizable magnetic field $B$, and since the electrons are tiny magnets, they interact with this field, and this interaction can distort the spectrum. Primarily, this distortion is noticable in the so-called **Zeeman effect**, that we now describe. 

In our derivation of the spectrum of the hydrogen atom, we concluded that the orbitals were labeled by their principal quantum number $n=1,2,3, ...$, but we did not consider the question of *how many electrons can fit in one orbital*. This number $d_n$ can be higher than one. If it is, we say that the orbital $n$ is **degenerate** and has **degeneracy** $d_n$. Since all electrons in a single orbital $n$ have the same energy $E_n=E_1/n^2$, specifying the energy of an electron does not uniquely determine its state.
