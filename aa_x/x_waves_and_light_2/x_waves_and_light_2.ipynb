## Optics

Simple waves are often described by the geometry of their wavefronts. The wavefronts in your 2D Desmos simulation are circular, so we might describe these as circular waves. In 3D, however, the analogous wave would be spherical, so these are usually called spherical waves. The simplest geometries are planar, so we might consider making plane waves. In a 2D simulation, the wavefronts would look like parallel lines.

Above, we emphasized that the speed of a wave depends not on the wavelength or frequency but on the *medium* through which the wave propagates. For example, sound waves in air are about 4 times slower than they are in water.   

An oft-repeated statement we have all heard is that the speed of light is fixed to be 	
$$
c=2.99792458108\ m/s	\tag{4}
$$
The correct statement is that any inertial observer who measures the speed of light *in vacuum* will find it to have this magnitude. When we say “nothing can go faster than the speed of light”, we really mean that no information can be transferred faster than the speed of light *in vacuum*. When light passes through transparent media like water or glass, it slows down. Other particles can travel faster through these media (as long as their speed does not exceed $c$), but this will produce shock waves of light called [Cherenkov radiation](https://en.wikipedia.org/wiki/Cherenkov_radiation). 

For an optically transparent medium, we define the **index of refraction**: instead of working directly with the speed of light in that medium $v\leq c$, we work with the ratio
$$
		n=\frac cv \quad \in[1,\infty)	\tag{5}
$$
of the speed of light in vacuum to the speed in the medium. Note that $n=1$ only in free space and is larger than that for any other transparent medium. The strange name comes from one of the hallmarks of waves: they **refract**. Refraction happens when a wave is incident on a medium that is transparent to that wave. As the wave passes into the medium, it changes speed because the medium sets the speed of the wave. If the wave slows down, by equation (1), its wavelength will get smaller. If the wavelength is incident on a medium/medium boundary at some angle, continuity of the wavefronts will force the wave to turn, as indicated in the figure at right.

Refraction is an important property of waves that has no analog for particles. Other important such properties are **diffraction** and **interference**. Similarly to reflection and refraction, diffraction is a type of change in the direction of motion as a wave encounters an obstacle with an associated size that is small relative to the wavelength of the wave. There are two main cases: The first is that the obstacle is a small object like a buoy or a piling is to an ocean wave. In this case the wave easily bends around the obstacle. In the second case, the obstacle is very large but there is an opening with a size small or comparable to the wavelength. When this happens, the wave “fans out” when it passes through the opening, as in the picture to the right.

![diffraction_1](./images/diffraction_1.png)

The specific patterns that you get from waves interacting with obstacles can be understood from [Huygens’ Principle](https://en.wikipedia.org/wiki/Huygens%E2%80%93Fresnel_principle), which states that for a propagating wave, the next wavefront will have the same shape as you would get from treating each point on the wave front as a source of spherical waves and taking the next wavefront to be the “envelope” of the spherical wavefronts. The schematic diagram below depicts this for the wave in the picture above. Note that as the opening becomes smaller, the resulting envelopes look more spherical. 

![diffraction_2](./images/diffraction_2.png)

Finally, interference is the phenomenon that two waves of the same type will change each other’s shape when they overlap. For simple waves, as we have been discussing here, the interference of two waves is gotten by simply adding them together. For waves that displace a medium like water or sound waves, this just means that we sum up the displacements. Since displacements can be both positive and negative, this means that the effect is sometimes to amplify the wave and other times to (partially) cancel out. We say that the waves are interfering **constructively** or **destructively**, respectively.

Now suppose we are looking through a telescope of aperture size D at two objects in the sky with an angular separation $\alpha$. According to problem 16, the light from each object will make an Airy diffraction pattern, so we see something like one of the cases in the above image. It becomes clear that once the central disks in the Airy patterns start to merge, we will lose the ability to distinguish a single source from a double source. Determining the **diffraction-limited resolution** of a telescope in this way is called the *Rayleigh criterion*. By definition, this is a limit on the resolving power of the telescope, and it depends only on the wavelength of the light as it compares to the aperture size of the telescope: 
$$
\theta_{min} \approx \lambda/D
$$

Actually, the pattern in problem 16 is not *exactly* (a cross section of) the Airy disk, but it is very close. Reanalyzing the case of a circular 2D aperture (rather than the 1D from problem 16\) results in the what is known as the Airy pattern which has slightly displaced maxima/minima. For example, the Rayleigh criterion reported [here](https://en.wikipedia.org/wiki/Angular_resolution#The_Rayleigh_criterion) is $\theta_{min} \approx 1.22\lambda/D$. 

We close by emphasizing that this diffraction limitation is the *best-case scenario*, not a problem that can be overcome with better technology (unless it somehow changes the apperture size). Any real telescope will *at best*, be limited by this fundamental behavior of waves as opposed to, for example, limits on the precision of the manufacturing process. To express this, we say that a telescope is **diffraction-limited** to a resolution of $\theta_{min} \approx \lambda/D$.

## The Planck Spectrum and Wien's Law

<a id="planck">
    
If you took AA:SS, we already discussed the **Planck curve** for a perfect emitter (a.k.a. a blackbody) in our study of the radiation spectrum of the Sun. In this class, we will discuss the phyics behind this curve in much more detail, but for now just try to follow the meaning behind the claims. 

When Planck was beginning his university studies in physics, he was told by his professor that physics was a highly mature field with only some details left to be figured out. Perhaps one of those "details" has become known at the [ultraviolet catastrophe](https://en.wikipedia.org/wiki/Ultraviolet_catastrophe): classical calculations in electromagnetism and thermodynamics predict that at small wavelengths an object in thermodynamics equilibrium *at any non-zero temperature* would radiate an infinite about of small-wavelength radiation. If you are thinking "what's the big deal?" please reread that as "the universe would be destroyed by a cold speck of dust emitting an infinite amount of gamma radiation" to say nothing of bigger hotter things. Clearly this is catastrophic for our understanding of physics. 

Planck did the thing that any professional physicist would do: he introduced a (positive) fudge factor he called $h$ with the hope that-at the end of the calculation-he could take $h\to 0$ and get the "correct" answer. In modern terms, the radiation curve he derived by introducing this factor is called the **spectral radiance**[^2.5](#fn) <a id="specrad">
$$
F_\lambda(T)=\frac{2\pi h c^2}{\lambda^5} \frac 1{e^{\frac{hc}{\lambda k_BT}} -1 }		\tag{Planck}
$$
Here, 
- $F_\lambda(T)$ is notation meant to indicate that this curve is a function (only) of the temperature $T$ of the object that is radiating light at wavelength $\lambda$. It has dimensions $M\cdot L^{-1}\cdot T^{-3}$, as we will verify in problem 14.
- $h$ is Planck's fudge factor.
- $c$ is the speed of light in vacuum. 
- $T$ is the temperature of the body emitting radiation. It does not matter what units $T$ is in becuase it is multiplied by the Boltzmann constant $k_B$, so $k_B T$ has dimensions of energy irrespective of the units chosen for temperature.
- Roughly speaking, $F_\lambda(T)$ is luminosity that an object of temperature $T$ is emitting at the wavelength $\lambda$. Strictly speaking, this does not make sense because a *single* wavelength (with infinite precision) contributes only an infinitesimal amount of power, so what it really means is this: consider the amount of power that an object at temperature $T$ is emitting at all wavelengths between $\lambda$ and $\lambda +d\lambda$ for some small $d\lambda \ll \lambda$. Then the amount of radiant flux the object is emitting in this "band" is $F_\lambda d\lambda$. We will use this later to determine the total radiant flux over all wavelengths $F:=\int_0^\infty F_\lambda d\lambda$.

### The Stefan-Boltzmann Law

As described above, the Planck curve gives a quantitative measure for the amount of electromagnetic radiation a blackbody emits in a band between $\lambda$ and $\lambda+d\lambda$. The **Stefan-Boltzmann Law**, by contrast, Says that the total flux emitted is given by 
$$
F= \sigma_B T^4
\quad \quad\textrm{where}\quad \quad
\sigma_B = \frac{\pi^2 k_B^4}{60 \hbar^3 c^2}	\tag{Stefan-Boltzmann Law}
$$
is called the **Stefan-Boltzmann constant**. Note the characteristic dependance on the temperature to the 4th power: A body at twice the temperature will have a luminosity 16 times higher!

In the next problem, we will integrate the spectral flux density $F_\lambda$ over all $\lambda\in[0,\infty)$ to get the total spectral flux $F$. In the non-calculus version, we will simply us the fact that desmos will calculate the integral numerically for you. However, it is known how to do the required integral in so-called "closed form". In the calculus version of the problem, we will set up the integral (but we need techniques from Dr. Osborne's AMT class to really learn how to get the value).
