## Waves

As astronomers, the information we get from the Universe comes to us almost exclusively in the form of light. Since almost everything we know about the cosmos is gleaned from the light we receive, we need to understand light and the way it interacts with matter in the most profound detail we can muster. 

Light is a wave disturbance of the **electromagnetic field**. Freely-propagating electromagnetic waves are also known as **radiation**. To understand what these abstract concepts mean, we will first look at more tangible and familiar examples of waves. Some of the most familiar waves are surface waves of water. Later in life, we may notice waves on strings and we learn that sound is a pressure wave. All the simpler waves have something that they disturb as they **propagate**. This material is called the **medium**. Water waves disturb what would otherwise be a flat and horizontal surface between air and water. Waves on a stretched string disturb the straight resting position of the string. In matter, atoms and molecules are separated from one another by distances that are very close to their average separation. Sound waves are disturbances of these atoms and molecules away from these average separations. 

Most physical waves are very complicated. For example, water waves near a beach change speed and shape as they move and often crest and crash into surf as they approach the shore. However, there are also very simple waves that are quite regular. The simplest waves undulate (crest and trough) in regular intervals in time called the **period** $P$ of the wave. You can make such a wave in a bathtub by taking a ruler and bobbing it up and down regularly at the surface of the water. Straight wave fronts parallel to the ruler will move away from the ruler at a constant speed, and–provided you are bobbing at regular intervals–the distances between successive peaks will be the same for all successive peaks. This distance is called the **wavelength** $\lambda$ (lambda) of the wave.

The reciprocal of the period $f=1/P$ is called the **frequency** of the wave, but it is often more convenient to define the **angular frequency** $\omega=2\pi f=\frac{2\pi}P$. The regular bobbing that you are doing with the ruler is called **simple harmonic motion**. An object undergoing simple harmonic motion is also said to be undergoing regular **oscillation**. If there is a rubber duckie floating on the water then, when the wave reaches it, it will start to oscillate regularly. How much it bobs up and down is called the **amplitude** $A$ of the wave. Note that the *amplitude is independent of the wavelength*: the wavelength is the distance between successive crests and the amplitude is the difference between the height of a crest and a trough.

In problem 1e, you determined that the speed of the wave is related to the period and wavelength by $v=\lambda f$ where $f=1/P$ is the frequency. Although there is nothing incorrect about your result, I find it misleading to leave it as written. As we have discussed, the speed $v$ of the wave is *set by the medium*, not by the relation between wavelength and period. Instead, I prefer to write the formula in this way:
<a tag="dispersion">
$$
\lambda =v \frac 1f \quad\quad\textrm{or} \quad\quad f=v \frac1\lambda
\tag{1}
$$
This way of writing the formula emphasizes that the wavelength of a wave is determined by its frequency (or frequency in terms of wavelength). Such a formula is called a **dispersion relation**.   

To summarize, a simple wave $\Psi(x,t)$ in one space dimension $x$ and one time $t$ can be put in the form   
<a id="wave">
$$
\Psi(x,t) =A\cos (kx - \omega t) \tag{2}
$$
This is a slight rewriting of your result in problem 1. In this expression:
* $A$ is the amplitude.  
* $\omega=2\pi f=2\pi /P$ is the aforementioned angular frequency. The 2 is there because the periodicity of the cosine is $2\pi$.  
* Similarly, $k=2\pi/\lambda$ is called the **wave number**. Up to a factor of $2\pi$, it is the reciprocal of the wavelength.
* The function $\Psi(x,t)$ is called the **wave function**.

## Intensity *vs.* pitch: amplitude *vs.* frequency

An effect that is often mistaken for the Doppler effect in the model above is the *intensity* (loudness) of the sound rather than the *pitch* (frequency): When an object that is making a sound is far away from us, the sound has a lower intensity than when it is closer to us. Suppose the noisy object is emitting sound in all directions evenly. This means that as the sound propagates outward, it is being spread evenly over a larger surface area. If the intensity of the emitter is constant at the source, that means that the intensity perceived as we move away from the source should fall off as $1/\textrm{distance}^2$. We can refine our model so that it takes this new effect into account as well.

```python
from aa_tools.astroconst import spectrum_wavelength
from astropy.units import nm, m, km, s
from astropy.constants import c

_lamda=560*nm
_lamda_prime=460*nm

print(f"\na.     From the figure, a wavelength of {_lamda} is {spectrum_wavelength(_lamda)}.")
print(f"\nb.     From the figure, a wavelength of {_lamda_prime} is {spectrum_wavelength(_lamda_prime)}.")

redshift = lambda x, x_prime: (x_prime/x-1)*c

_blueshift=redshift(_lamda.to(m), _lamda_prime.to(m))

print(f"\nb iii. An object emitting at wavelength {_lamda} but apparent wavelength {_lamda_prime} has a\n       radial velocity of { (_blueshift.to(km/s)).value:.1e} {(_blueshift.to(km/s)).unit}.")
print(f"       This is {round(_blueshift/c, 2)}c.\n")
```

In the problem above, the light’s wavelength is shifted to the shorter side. In the visible part of the spectrum, blue light is on the short side (450 – 485 `nm`) whereas red light is on the longer side (625 – 750 `nm`). Because of this, we refer to light that is Doppler shifted to shorter apparent wavelengths as **blue-shifted** and light that is Doppler shifted to longer apparent wavelengths as **red-shifted**. We use this terminology not just for light, but for other waves as well. 

Note that the Doppler shift we have been discussing can only tell us about the part of the relative motion that is directly toward/away from us. This type of motion is called **radial motion**. (The transversal component of the motion is called **tangential motion** (because this component is tangent to the celestial sphere), but the Doppler shift as we have discussed it, does not tell us anything about that part.)

## Optics

Simple waves are often described by the geometry of their wavefronts. The wavefronts in your 2D Desmos simulation are circular, so we might describe these as circular waves. In 3D, however, the analogous wave would be spherical, so these are usually called spherical waves. The simplest geometries are planar, so we might consider making plane waves. In a 2D simulation, the wavefronts would look like parallel lines.

Above, we emphasized that the speed of a wave depends not on the wavelength or frequency but on the *medium* through which the wave propagates. For example, sound waves in air are about 4 times slower than they are in water.   

An oft-repeated statement we have all heard is that the speed of light is fixed to be 	
$$
c=2.99792458108\ m/s	\tag{4}
$$
The correct statement is that any inertial observer who measures the speed of light *in vacuum* will find it to have this magnitude. When we say “nothing can go faster than the speed of light”, we really mean that no information can be transferred faster than the speed of light *in vacuum*. When light passes through transparent media like water or glass, it slows down. Other particles can travel faster through these media (as long as their speed does not exceed $c$), but this will produce shock waves of light called [Cherenkov radiation](https://en.wikipedia.org/wiki/Cherenkov_radiation). 

For an optically transparent medium, we define the **index of refraction**: instead of working directly with the speed of light in that medium $v\leq c$, we work with the ratio
$$
		n=\frac cv \quad \in[1,\infty)	\tag{5}
$$
of the speed of light in vacuum to the speed in the medium. Note that $n=1$ only in free space and is larger than that for any other transparent medium. The strange name comes from one of the hallmarks of waves: they **refract**. Refraction happens when a wave is incident on a medium that is transparent to that wave. As the wave passes into the medium, it changes speed because the medium sets the speed of the wave. If the wave slows down, by equation (1), its wavelength will get smaller. If the wavelength is incident on a medium/medium boundary at some angle, continuity of the wavefronts will force the wave to turn, as indicated in the figure at right.

Refraction is an important property of waves that has no analog for particles. Other important such properties are **diffraction** and **interference**. Similarly to reflection and refraction, diffraction is a type of change in the direction of motion as a wave encounters an obstacle with an associated size that is small relative to the wavelength of the wave. There are two main cases: The first is that the obstacle is a small object like a buoy or a piling is to an ocean wave. In this case the wave easily bends around the obstacle. In the second case, the obstacle is very large but there is an opening with a size small or comparable to the wavelength. When this happens, the wave “fans out” when it passes through the opening, as in the picture to the right.

![diffraction_1](./images/diffraction_1.png)

The specific patterns that you get from waves interacting with obstacles can be understood from [Huygens’ Principle](https://en.wikipedia.org/wiki/Huygens%E2%80%93Fresnel_principle), which states that for a propagating wave, the next wavefront will have the same shape as you would get from treating each point on the wave front as a source of spherical waves and taking the next wavefront to be the “envelope” of the spherical wavefronts. The schematic diagram below depicts this for the wave in the picture above. Note that as the opening becomes smaller, the resulting envelopes look more spherical. 

![diffraction_2](./images/diffraction_2.png)

Finally, interference is the phenomenon that two waves of the same type will change each other’s shape when they overlap. For simple waves, as we have been discussing here, the interference of two waves is gotten by simply adding them together. For waves that displace a medium like water or sound waves, this just means that we sum up the displacements. Since displacements can be both positive and negative, this means that the effect is sometimes to amplify the wave and other times to (partially) cancel out. We say that the waves are interfering **constructively** or **destructively**, respectively.

Phew! That was a lot. In the last few problems, we worked hard to arrive at the conclusion that for a wave of wavelength $\lambda$ passing through an opening of size $D$, we can expect diffraction of the wave of angular size set by the ratio $\lambda/D$ as you derived in equation (7).

Now suppose we are looking through a telescope of aperture size D at two objects in the sky with an angular separation $\alpha$. According to problem 16, the light from each object will make an Airy diffraction pattern, so we see something like one of the cases in the above image. It becomes clear that once the central disks in the Airy patterns start to merge, we will lose the ability to distinguish a single source from a double source. Determining the **diffraction-limited resolution** of a telescope in this way is called the *Rayleigh criterion*. By definition, this is a limit on the resolving power of the telescope, and it depends only on the wavelength of the light as it compares to the aperture size of the telescope: 
$$
\theta_{min} \approx \lambda/D
$$

Actually, the pattern in problem 16 is not *exactly* (a cross section of) the Airy disk, but it is very close. Reanalyzing the case of a circular 2D aperture (rather than the 1D from problem 16\) results in the what is known as the Airy pattern which has slightly displaced maxima/minima. For example, the Rayleigh criterion reported [here](https://en.wikipedia.org/wiki/Angular_resolution#The_Rayleigh_criterion) is $\theta_{min} \approx 1.22\lambda/D$. 

We close by emphasizing that this diffraction limitation is the *best-case scenario*, not a problem that can be overcome with better technology (unless it somehow changes the apperture size). Any real telescope will *at best*, be limited by this fundamental behavior of waves as opposed to, for example, limits on the precision of the manufacturing process. To express this, we say that a telescope is **diffraction-limited** to a resolution of $\theta_{min} \approx \lambda/D$.
