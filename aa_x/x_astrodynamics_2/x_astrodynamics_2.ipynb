# Gravitational Potential Energy
Energy is an important physical concept that can be used in many situations instead of forces, and–when applicable–it is often much easier to use. The basic reason for this is that whereas the force ${\bf F}$ is a vector quantity, the energy is just an ordinary function.

Energy comes in very many forms, but the most important to us are the energy associated with motion and the energy associated with forces.

Potential energy is the energy associated with forces. A force on an object will try to move that object in such a way as to lower its potential energy. It is probably easiest to work directly with examples. The most relevant example for us is the gravitational force given by the LUG
$$
F_{g} = G\frac{m_{1}m_{2}}{r^{2}} \tag{1}
$$
The gravitational potential energy function associated with this is denoted by Ug(r) and is given by
$$
U_{g}(r) = {- G}\frac{m_{1}m_{2}}{r} \tag{2}
$$

Note two important differences between formulae (1) and (2): There is a minus sign and the power of r in the denominator is different.[1]

Let’s use the results of this exercise to calculate $\Delta U/\Delta r$. When we move from a point $r$ to a nearby point $r + \Delta r$, the gravitational potential energy changes from $U_g(r)$ to $U_g(r + \Delta r)$, so the change in potential energy is $\Delta U_g = U_g(r + \Delta r) − U_g(r)$. Saying that the second point is nearby means that $\Delta r$ is very small compared to $r$, or $0 <\frac{\Delta r}{r} \ll 1$, just as we required from $ε$ in problem 2.

Now we write
$$
U_g(r+\Delta r)=-G \frac {m_1m_2}{r+\Delta r}
=-G\frac {m_1m_2}r \frac1{1+\varepsilon}
=U_g(r)\frac1{1+\varepsilon}
$$
where we have defined $\varepsilon:= \frac {\Delta r}r$. 
>The notation "$:$" is used in mathematics to mean "such that". Thus "$:=$" can be read as "such that it equals", and the whole statement becomes "...defined $\varepsilon$ such that it equals $\frac {\Delta r}r$."

Next, we expand in $\varepsilon$ using our result from problem 2:
$$
U_g(r+\Delta r)=-G\frac {m_1m_2}r \frac1{1+\varepsilon}
=U_g(r) \frac1{1+\varepsilon}
=U_g(r) \left[ 1-\varepsilon+ \varepsilon^2- \varepsilon^3+\dots \right]
$$
Thus $\Delta U_g(r)=U_g(r+\Delta r)-U_g(r)$ gives 
$$
\Delta U_g(r)=\Delta r\left(-\frac1r\right) U_g(r) \left[ 1-\varepsilon+\dots \right]
$$
So $\frac{\Delta U_g(r)}{\Delta r}$ is very nearly equal to $\left(-\frac1r\right) U_g(r)$ with the approximation getting better and better as $\varepsilon$ gets smaller and smaller. In fact, we can safely take $\varepsilon \to 0$ now and get an exact answer! This is the definition of the derivative that you will learn when you take (pre)calculus:
$$
\frac{d U_g(r)}{dr}=\lim_{\varepsilon\to 0} \frac{\Delta U_g(r)}{\Delta r}
\quad \quad (\textrm{derivative})
$$
and thus, we have derived that 
$$
\frac{d U_g(r)}{dr}=-\frac1r U_g(r)=+\frac1r G\frac{m_1m_2}r= G\frac{m_1m_2}{r^2}=F_g(r)
$$
proving the claim made in formula (3): We have recovered the gravitational force from the gravitational potential energy!

# Mechanical Energy

The mechanical energy $E$ of a system is the sum of the kinetic energy $K=\frac12 mv^2$ and the potential energy $U$: 
$$
E=K+U(r)=\tfrac 12 mv^2 + U(r)
$$
When you take physics, you will learn a deep and wonderful fact about the way that Nature works: if no external influences are bothering your system (a.k.a. "isolated system"), then the mechanical energy of the system is **conserved**, which means it will have a constant value that does not change in time.

This is wonderful because even though the kinetic energy $K$ may change and the potential energy $U(r)$ may change, the sum of the two will not. Let’s see how this helps us in a typical physics example.

In problem 3d, you showed that for an object of mass $m$ a height $h$ above the ground, the gravitational potential energy is $U_g(h) = mgh$ where $g=\frac{GM_\oplus}{R_\oplus^2}$ is the usual surface gravity of $9.8 m/s^2$. Now suppose we drop this object from rest. How fast is it going when it hits the ground? We will use the fact that the energy is conserved:  
The total mechanical energy when the object is at height $h$ must be equal to that at height 0. Since the object is dropped from height $h$ from rest, $v=0$ here:
$$
E_h=\frac12 mv^2+mgh=mgh \quad \quad \textrm{dropped from rest at height $h=0$.}
$$
Now, at height $h=0$
$$
E_0=\frac12 mv^2+mgh=E=\frac12 mv^2 \quad \quad \textrm{since $h=0$.}
$$
Conservation of energy means that these two energies need to be equal $E_h=E_0$, so
$$
\frac12 mv^2 = mgh \quad \Rightarrow \quad v^2 = 2gh
$$
The question was what the speed of the object is when it gets to the ground, so now we solve for $v$. The masses cancel out so all we had to do was multiply by 2. Now we take the square root: $v=\sqrt{2gh}$.

Note that the result above assumes that $h\ll R_\oplus$. We will generally be interested in systems where the only relevant force is the gravitational influence of a massive body, say of mass $M$. For general distances $r$ not bounded by $R_\oplus$, we need to go back to the general formula for the gravitational energy: 
$$ 
E=\frac12 mv^2-G\frac {mM}{r}  \tag{4}
$$

This energy is again constant, so we can play the same game we played in the example.

## Escape speed

If you throw a ball straight upward, it will slow to a stop, turn around, and come back down. If you throw it harder, it will go higher before it comes back down. If you throw even harder, it will go higher still. Is there a speed you can throw it so that it will not come back?

A ball with mass $m$ on Earth has some gravitational potential energy given by the formula (2) with $r=R_\oplus$ and the other mass equal to the Earth’s mass $M_\oplus$. If that same ball is moved infinitely far away from the Earth, it will have 0 gravitational potential energy. Instead of moving it out infinitely far away, we want to try to throw it there. 

If the ball launched from Earth with speed $v$, the total energy (4) must be the same when it is launched as when it arrives at infinity. When it arrives at infinity it may still have some excess speed, but the minimum initial speed that will get it there will have the ball arriving with no excess kinetic energy. That minimum initial speed is called the **escape speed** $v_e$: anything traveling at that speed or faster will escape the gravitational field of the Earth.

## Center of mass

Newton made a correction to Kepler’s First Law: What goes at the focus is not the primary body itself but the **center of mass** of the primary and secondary together. This center of mass is basically the correct generalization of the notion of average when the objects contribute different amounts. It's like your grade: It depends on formative assessments and summative ones, but the former count for $30\%$ and the latter by $70\%$, so we don't just add the scores. Instead we take an average weighted by these numbers. Suppose your formative score is denoted by $f$ and your summative by $s$. Then instead of grade $= \frac{1\times f+1\times s}2$, we calculate that it is $=\frac{30\times f+70 \times s}{30+70}$. In this analogy, the mass $M$ of the primary is the $70\%$, that $m$ of the secondary is $30\%$, and the total mass $M+m$ is $100\%$.

Now, suppose you have a very light (negligible mass) meter stick with a 1 kg mass attached to each end. Then you expect the center of mass to be at the center. But if we have a 100 kg mass at one end and a 1 kg mass at the other, we expect the center of mass to be much closer to the 100 kg mass.

In general, if two masses $m_1$ and $m_2$ are on a line with coordinates $x_1$ and $x_2$ respectively, then the **center of mass** or **barycenter** of the system is at
$$
x_{cm}=\frac 1M \left[m_1 x_1 + m_2 x_2\right] \tag{CM}
$$
where $M=m_1+m_2$ is the total mass of the system.

The structure of this formula is the following: The term in brackets does the “averaging” over the values of $x$ weighted by the mass: If there is no mass at some value of $x_*$ then the weight is 0 so there is no contribution to the sum there. Since $m_1$ is at $x_1$, we get a term $m_1 x_1$ for the first mass, and similarly for $m_2$ at $x_2$. The term in brackets has dimensions of mass times length, so to get something with dimensions of length, we need to divide by something with dimensions of mass. That thing will be the total mass $M=m_1 + m_2$. which is unbiased with respect to 1 and 2. Note that if $m_1 = m_2$, then 
$M=m_1+m_2$, and both masses contribute equally and the average reduces to the average of the position

## From one-body to two-body and back

Kepler’s Laws are correct in the so-called one-body approximation. In this approximation the Sun is not dynamical. Instead, it is fixed at the center $r=0$, and the planet that orbits it is the only dynamical object, ergo one-body.

In contrast to this, Newton treated the Sun as dynamical too, ergo a two-body problem instead of a one-body problem. In the two-body problem, there are two dynamical masses and one gravitational interaction between them. The conserved energy for this problem is
$$
E =\tfrac12 m_1v_1^2+\tfrac12 m_2v_2^2 -G\frac{m_1m_2}r \tag{6}
$$
It consists of a kinetic energy term for each mass $m_1$ and $m_2$ separated by a distance $r$ and the gravitational potential energy between them. Following Newton, we can rewrite this expression as follows:
$$
E =\tfrac12 MV^2+\tfrac12 mv^2 -G\frac{mM}r \tag{6'}
$$
In this expression, $v$ is the relative speed between the two masses, $M = m_1 + m_2$ is the total mass of the system, $V$ is the speed of the center of mass, and
$$
m= \frac {m_1m_2}{m_1+m_2}
$$
is a combination called the reduced mass. Expression (6’) says that the energy of the original system can interpreted as arising from two independent parts:
- First, there is the kinetic energy of the center of mass of the whole system given by $\frac12 MV^2$
- Second, we recognize $\frac12 mv^2 -G\frac{mM}r$ from equation (4). It is the total energy of the one-body system consisting of the secondary of mass $m$ orbiting a fixed primary with mass $M$.

Thus what we have found is that the two-body problem can be reinterpreted as a one-body problem by going to the rest frame of the center of mass!

So what have we learned? The proper treatment of a gravitating system of two astronomical bodies should treat both as participating in the motion (two-body dynamics) instead of freezing the larger one in place and having only the second in motion (one-body dynamics). If we do this, then we find that the problem can be reduced from the more complicated two-body problem to the simpler one-body problem, provided we make the following adjustments
- We move along with the system at its center of mass.
- The what is fixed at the focus is a fictional body with mass $M = m_1 + m_2$ equal to the total mass of the system.
- What “orbits” is a fictional body with mass $m= \frac {m_1m_2}{m_1+m_2}$ equal to the reduced mass of the system.
- The distance $r$ between these fictional bodies is equal to the original distance between $m_1$ and $m_2$.

Although we will not do it here, a little more work shows how to take the solution of the motion of the reduced one-body system and translate it back to the original two-body problem:
- The original two bodies orbit their common center of mass in ellipses with semi-major axes $a_1$ and $a_2$. The semi-major axis of the ellipse of the reduced problem is related to these by $a = a_1 + a_2$.
- The eccentricities of all three ellipses (*i.e.* that of the primary, secondary, and reduced problem) are equal.
- The two original bodies orbit their center of mass with the same period. Both are equal to the period of the 1-body problem which was given by $\left(\frac{P}{2\pi}\right)^2 =\frac{a^3}{GM}$, but now $M = m_1 + m_2$ is the total mass of the system and a = a1 + a2 is the semi-major axis of the reduced problem.
